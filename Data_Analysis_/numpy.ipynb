```python
def mse(y_pred , y_true):
    y = y_pred - y_true
    y = y ** 2
    return mse.mean(y)
```

```python
mse(np.array([3, -0.5, 2, 7]) , np.array([2.5 , 0.0 , 2 , 8]))
```
