```python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive the {self.engine_type} car.")

    def stop(self):
        print("The car has stopped.")
```

```python
car_one = Car(4, 5, "petrol")
car_one.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_driving):
        super().__init__(windows, doors, engine_type)
        self.is_self_driving = is_self_driving

    def self_driving_feature(self):
        print(f"Tesla supports self driving: {self.is_self_driving}")
```

```python
tesla_one = Tesla(4, 5, "electric", True)
tesla_one.self_driving_feature()
tesla_one.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, battery_capacity):
        super().__init__(windows, doors, engine_type)
        self.battery_capacity = battery_capacity

    def battery_info(self):
        print(f"Tesla has a battery capacity of: {self.battery_capacity} kWh")
```

```python
class Toyota(Car):
    def __init__(self,windows , doors , engine_type , colour):
        super().__init__(windows , doors , engine_type)
        self.colour = colour

    def car_colour(self):
        print(f"Toyota car colour is: {self.colour}")
        
```

```python
car2 = Toyota(4,5 , "disel "  , " magic black ")
car2.car_colour()
car2.drive()
car2.stop()
```

```python
tesla_one = Tesla(4, 5, "electric", 100)
tesla_one.battery_info()
```

## AGGREGATION - when one class owns another class

```python
class Customer:
    def __init__(self , name , age , address):
        self.name = name 
        self.age = age
        self.address = address

    def get_customer_info(self):
        print(f"Customer Name: {self.name}, Age: {self.age}, Address: {self.address.street}, {self.address.get_city}, {self.address.zip_code}")

class Address:
    def __init__(self , street , city , zip_code):
        self.street = street
        self.__city = city
        self.zip_code = zip_code

    def get_city(self):
        return self.__city

address1 = Address("123 Main St" , "New York" , "10001")
customer1 = Customer("John Doe" , 30 , address1)
customer1.get_customer_info()
```

## Decorators 

takes another fucntions and add or decorate to them because of ptyhon fuctiosn are first class citizen 
builtin function = @staticmethod , @classmethod 
and other user defined 

```python
def my_decorator(func):
    def wrapper():
        print("**********")
        func()
        print("**********")
    return wrapper

def hello():
    print("Hello, World!")

def display():
    print("Display function called.")


decorated_hello = my_decorator(hello)
decorated_hello()

decorated_display = my_decorator(display)
decorated_display()
```
