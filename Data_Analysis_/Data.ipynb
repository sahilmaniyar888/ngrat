```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 1: Prepare data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Step 2: Build model
model = LinearRegression()

# Step 3: Train model
model.fit(x, y)

# Step 4: Predict
prediction = model.predict([[6]])
print("Prediction for x = 6:", prediction[0])
```
