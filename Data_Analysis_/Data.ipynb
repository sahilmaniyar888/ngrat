```python
import pandas as pd
```

```python
# series = one dimensional = coloumn 

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
```

```python
# series from dictonary

data = {'a': 10, 'b': 20, 'c':30}
series = pd.Series(data)
print(series)
```

```python
# dataframe
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}

df = pd.DataFrame(data)
print(df)
```
