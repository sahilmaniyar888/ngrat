```python
class Hero:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def display(self):
        print(f"Hero: {self.name}, Power: {self.power}")
```

```python
Hero("Superman", "Flight").display()
```

```python
n = 10
total = n * (n + 1) // 2
print("Sum =", total)
```

```python
n = 10
total = 0
i = 1

while i <= n:
    total += i
    i += 1

print("Sum =", total)
```
