```python
class Shape():
    def area(self):
        return " The area is undefined for generic shape"
    
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def area(self):
        return self.width * self.height
        
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return 3.14 * self.radius ** 2
        

```

```python
react = Rectangle(5, 10)
print("Area of Rectangle:", react.area())   
circ = Circle(7)
print("Area of Circle:", circ.area())
```
